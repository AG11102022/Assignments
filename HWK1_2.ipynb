## Simulate (sample from) the bivariate normal distribution with the shown parameters obtaining a plot similar to Figure 6.8b that shows the simulation result from a different bivariate Gaussian distribution. You can generate m = 200 samples/points (10 points)

```python
 # Importing the necessary modules
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import multivariate_normal
from scipy.linalg import svd
from sklearn.decomposition import PCA
import seaborn as sns

from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
import tensorflow as tf

from prml.preprocess import GaussianFeature, PolynomialFeature, SigmoidalFeature
from prml.linear import (
    BayesianRegression,
    EmpiricalBayesRegression,
    LinearRegression,
    RidgeRegression
)
```

### Parameters

```python
# Initializing the random seed
random_seed=1000

# Parameters
mean = np.array([0, 2])  # Mean
cov = np.array([[0.3, -1], [-1, 5]])  # Covariance matrix

# samples
num_samples = 200
```

### Gaussian bivariate distribution

```python
# Generating a Gaussian bivariate distribution
# with given mean and covariance matrix
distr = multivariate_normal(mean, cov, seed = random_seed)
```

### Generating samples

```python
# Generating 200 samples out of the
# distribution
data = distr.rvs(num_samples)
```

```python
### Extract coordinates
```

```python
 # x1 and x2
x1 = data[:,0]
x2 = data[:,1]
```

### Plotting

```python
# Plotting the generated samples
plt.figure(figsize=(8, 6))
plt.plot(x1, x2, 'o', c='purple', alpha = 0.7)
plt.title('Bivariate Gaussian Distribution')
plt.xlabel('x1')
plt.ylabel('x2')

# Set x and y axis limits
plt.xlim(-2, 2)
plt.ylim(-4, 8)

plt.show()
```

### Meshgrid

```python
# Define the range for x and y for the contour plot
margin = 0.5
x = np.linspace(np.min(x1) - margin, np.max(x1) + margin, num_samples)
y = np.linspace(np.min(x2) - margin, np.max(x2) + margin, num_samples)
X, Y = np.meshgrid(x, y)
```

# Problem 2: Projection (20 points)

### Simulate a 3-dimensional (3d) Gaussian random vector with the following covariance matrix by sampling m = 1000 3D vectors from this distribution.
### Using the Singular Value Decomposition (SVD) of the covariance matrix compute the projection of the m simulated vectors onto the subspace spanned by the first two principal components (or left singular vectors of the covariance matrix).

 


```python
cov = np.array([[4, 2, 1], [2, 3, 1.5], [1, 1.5, 2]])  # Covariance matrix
```

### SVD

```python
U, S, VT = svd(cov)
```

```python
print("U (Left Singular Vectors):")
print(U)
print("S (Singular Values):") # eigenvalues
print(S)
print("VT (Right Singular Vectors):")
print(VT)
```

### What determines the principal components? Show the vectors which denote the first 2 principal components

The principal components are represented by the columns of the left singular vector ( or U). The first two principal components are the first two columns of U

```python
# Extract the first two principal components
principal_components = U[:,:2]
print(principal_components)
```

### Direct PCA on covariance matrix (just to understand simple case)

### Plot the projected vectors in the subspace of first 2 principal components.

```python
# Perform PCA
pca = PCA(n_components=2)
projected_data = pca.fit_transform(cov)

print(projected_data)
```

```python
# Plot the projected data vectors
plt.figure(figsize=(8, 6))
# Plot the projected data vectors
for i, vector in enumerate(projected_data):
    plt.plot([0, vector[0]], [0, vector[1]])
    plt.text(vector[0], vector[1], f"Vector {i+1}", fontsize=12, ha='right', va = 'top')

# first two principal components using plt.plot
principal_components = pca.components_
plt.plot([0, principal_components[0, 0]], [0, principal_components[0, 1]], color='r', label='First Principal Component')
plt.text(principal_components[0, 0], principal_components[0, 1], 'PC1', color='r', fontsize=12)
plt.plot([0, principal_components[1, 0]], [0, principal_components[1, 1]], color='m', label='Second Principal Component')
plt.text(principal_components[1, 0], principal_components[1, 1], 'PC2', color='m', fontsize=12)

# Adjust plot limits
plt.xlim(-2, 2)
plt.ylim(-0.7, 1.2)

plt.title("Projection onto First 2 Principal Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.axhline(0, color='grey', linestyle='--', linewidth=0.5)
plt.axvline(0, color='grey', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

### Reverse the projection to map back to the original 3D space and create a scatter plot to show the reconstructed points. Do the reconstructed points have identical/similar but not identical/different correlations in respective components as the original matrix?

```python
# Reverse the projection to map back to the original 3D space
reconstructed_data = pca.inverse_transform(projected_data)

# Figure dimension
fig = plt.figure(figsize=(12, 6))

# original data points
ax = fig.add_subplot(121, projection='3d')
ax.scatter(cov[:, 0], cov[:, 1], cov[:, 2], color='r', label='Original Points')
ax.set_title("Original Data Points")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

# reconstructed data points
ax = fig.add_subplot(122, projection='3d')
ax.scatter(reconstructed_data[:, 0], reconstructed_data[:, 1], reconstructed_data[:, 2], color='b', label='Reconstructed Points')
ax.set_title("Reconstructed Data Points")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

plt.show()

#  correlations
original_corr = np.corrcoef(cov.T)
reconstructed_corr = np.corrcoef(reconstructed_data.T)

print("Original Correlation Matrix:")
print(original_corr)

print("\nReconstructed Correlation Matrix:")
print(reconstructed_corr)
```

# Problem 3: Stochastic Gradient Descent (30 points)

### In class we covered the baseline stochastic gradient descent. Using the linear regression example from the class notes, develop from scratch the baseline SGD algorithm. :
### Clearly state the hyper-parameters you used and present the loss vs epoch plot that demonstrates the convergence of the algorithm.

```python
# !pip install git+https://github.com/pantelis-classes/PRML.git#egg=prml
# !pip install C:\Users\algba\OneDrive\Desktop\AIcourse\PRML-master


# Apply the default theme
sns.set_theme()

np.random.seed(1234)

def create_toy_data(func, sample_size, std, domain=[0, 1]):
    x = np.linspace(domain[0], domain[1], sample_size)
    np.random.shuffle(x)
    y = func(x) + np.random.normal(scale=std, size=x.shape)
    return x, y

def sinusoidal(x):
    return np.sin(2 * np.pi * x)

x_train, y_train = create_toy_data(sinusoidal, 10, 0.25)
x_test = np.linspace(0, 1, 100)
y_test = sinusoidal(x_test)

plt.figure(figsize=[10,8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, label="$\sin(2\pi x)$")
plt.plot(x_test, y_test, "-g", label="Target Function")
plt.legend()
plt.show()
```

# Linear SGD

```python
# From linear regression example in class
x_train = x_train.reshape(-1,1)
y_train = y_train.ravel()

x_test = x_test.reshape(-1,1)
y_test = y_test.ravel()

# Using SGD
sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, learning_rate='invscaling', eta0=0.01, random_state=42)
sgd_reg.fit(x_train, y_train)

# Prediction
y_pred = sgd_reg.predict(x_test)

# Plot the results
plt.figure(figsize=[10, 8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, "-g", label="Target Function")
plt.plot(x_test, y_pred, "-r", label="SGD Prediction")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("SGD for Linear Regression using Scikit-Learn")
plt.show()

# Epoch vs loss

max_iter=1000


loss_values = []


for epoch in range(max_iter):
    sgd_reg.partial_fit(x_train, y_train)
    y_pred = sgd_reg.predict(x_train)
    mse = np.mean((y_pred - y_train) ** 2)
    loss_values.append(mse)

# Plot the loss versus epoch
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_iter + 1), loss_values, color='blue')
plt.title('Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
# Set random seed for reproducibility
np.random.seed(1234)

# hyperparameters
learning_rate = 0.01
epochs = 1000
degree = 5  

x_train, y_train = create_toy_data(sinusoidal, 10, 0.25)
x_train = x_train.reshape(-1, 1)  # Reshape to column vector


poly = PolynomialFeatures(degree)
scaler = StandardScaler()

x_train_poly = poly.fit_transform(x_train)
x_train_poly = scaler.fit_transform(x_train_poly)

sgd_regressor = SGDRegressor(max_iter=1, tol=None, eta0=learning_rate, learning_rate='constant', random_state=1234, warm_start=True)


loss_history = []


for epoch in range(epochs):
    sgd_regressor.fit(x_train_poly, y_train)
    y_train_pred = sgd_regressor.predict(x_train_poly)
    loss = mean_squared_error(y_train, y_train_pred)
    loss_history.append(loss)

    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss:.4f}")

# loss vs. epoch
plt.figure(figsize=[10, 6])
plt.plot(loss_history, label="Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss vs. Epoch")
plt.legend()
plt.show()

# Generate test data
x_test = np.linspace(0, 1, 100).reshape(-1, 1)
y_test = sinusoidal(x_test)
x_test_poly = poly.transform(x_test)
x_test_poly = scaler.transform(x_test_poly)

# Predictions
y_pred_sgd = sgd_regressor.predict(x_test_poly)

# Plot training data, target function, and SGD polynomial predictions
plt.figure(figsize=[10, 8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="Training Data")
plt.plot(x_test, y_test, "-g", label="Target Function")
plt.plot(x_test, y_pred_sgd, "-r", label="SGD Polynomial Prediction")
plt.legend()
plt.title("Training Data, Target Function, and SGD Polynomial Prediction")
plt.show()
```

### Prof method - from chapter 4

```python
np.random.seed(1234)

#  hyperparameters
n_epochs = 50
t0, t1 = 5, 50  # learning schedule hyperparameters
degree = 5  

# Create toy training data
x_train, y_train = create_toy_data(sinusoidal, 10, 0.25)
x_train = x_train.reshape(-1, 1)  # Reshape to column vector

# Polynomial features and standardization
poly = PolynomialFeatures(degree)
scaler = StandardScaler()

x_train_poly = poly.fit_transform(x_train)
x_train_poly = scaler.fit_transform(x_train_poly)

#  bias term
X_b = np.c_[np.ones((x_train_poly.shape[0], 1)), x_train_poly]

theta = np.random.randn(X_b.shape[1], 1) 


def learning_schedule(t):
    return t0 / (t + t1)

m = len(X_b)
loss_history = []

for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y_train[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients

    
    y_train_pred = X_b.dot(theta)
    loss = mean_squared_error(y_train, y_train_pred)
    loss_history.append(loss)

print(theta)

# loss vs. epoch plot
plt.figure(figsize=[10, 6])
plt.plot(range(n_epochs), loss_history, label="Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss vs. Epoch")
plt.legend()
plt.show()

# Generating test data
x_test = np.linspace(0, 1, 100).reshape(-1, 1)
y_test = sinusoidal(x_test)
x_test_poly = poly.transform(x_test)
x_test_poly = scaler.transform(x_test_poly)
X_b_test = np.c_[np.ones((x_test_poly.shape[0], 1)), x_test_poly]

# Make predictions using the trained polynomial model
y_pred = X_b_test.dot(theta)

# Plot training data, target function, and SGD polynomial predictions
plt.figure(figsize=[10, 8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="Training Data")
plt.plot(x_test, y_test, "-g", label="Target Function")
plt.plot(x_test, y_pred, "-r", label="SGD Polynomial Prediction")
plt.legend()
plt.title("Training Data, Target Function, and SGD Polynomial Prediction")
plt.show()
```

### Tensor flow and Adam optimizer

```python
# model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(x_train_poly.shape[1],)),
    tf.keras.layers.Dense(1)
])

# Compile the model with Adam optimizer
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='mse')

# Train model
history = model.fit(x_train_poly, y_train, epochs=1000, verbose=0)

# Plot loss vs. epoch
plt.figure(figsize=[10, 6])
plt.plot(history.history['loss'], label="Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss vs. Epoch")
plt.legend()
plt.show()

# Predictions
y_pred_adam = model.predict(x_test_poly)

# Plot training data, target function, and Adam polynomial predictions
plt.figure(figsize=[10, 8])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="Training Data")
plt.plot(x_test, y_test, "-g", label="Target Function")
plt.plot(x_test, y_pred_adam, "-r", label="Adam Polynomial Prediction")
plt.legend()
plt.title("Training Data, Target Function, and Adam Polynomial Prediction")
plt.show()
```
